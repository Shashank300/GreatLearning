### Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference

```python
def Goal_diff_count():
    laliga['Goal_diff_count']=laliga.GoalsFor.astype(int)-laliga.GoalsAgainst.astype(int)
    
    return(laliga[['Team','Goal_diff_count']].sort_values(by='Goal_diff_count',ascending=False))
Goal_diff_count()
    
```

```python
#Team with the maximum goal difference

print(Goal_diff_count().head(1))
```

```python
#Team with the minimum goal difference

print(Goal_diff_count().tail(1))
```

### 5. Create a new column with name “Winning Percent” and append it to the data set (5 points)

#### Percentage of Winning = (GamesWon / GamesPlayed)*100

#### If there are any numerical error, replace it with 0%

#### Print the top 5 teams which has the highest Winning percentage

```python
laliga['Winning Percent']=round((laliga['GamesWon'].astype(int)/laliga['GamesPlayed'].astype(int)*100),2)
```

```python
#Replacing numerical error with 0
laliga.loc[laliga['Winning Percent'].isnull(),'Winning Percent']=0
```

```python
#team with higgest winning percentage

print(laliga[['Team','Winning Percent']].sort_values(by='Winning Percent',ascending=False).head(5))
```
